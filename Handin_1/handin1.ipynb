```python
import pandas as pd

data = pd.read_csv('https://wfs-kbhkort.kk.dk/k101/ows?service=WFS&version=1.0.0&request=GetFeature&typeName=k101:monumenter&outputFormat=csv&SRSNAME=EPSG:4326')
print(data.head())
#data1 = data.loc[data['id'] == 49690]
#print((data1.iloc[0,7], data1.iloc[0,8]))
```

## Opgave 2

```python
vedligeholdt = data.loc[data['graffitirenhold'] == 'Ja']
print(f'Vedligeholdte monumenter: {len(vedligeholdt)}')
```

## Opgave 3

```python
def monument_location(id=None, name=None):
    if id != None:
        row = data.loc[data['id'] == id]
    else:
        row = data.loc[data['navn'] == name]
    lat, lon = row.iloc[0,8], row.iloc[0,7]
    return lat, lon

print(monument_location(id=50063))
print(monument_location(name='Efterår'))
```

## Opgave 3.a

```python
import folium

def monument_map(pos):
    map_osm = folium.Map(location=(pos[0], pos[1]), zoom_start=13)
    folium.Marker((pos[0], pos[1])).add_to(map_osm)
    map_osm.save('./handinosm.html')
    map_osm
    
monument_map(monument_location(id=49866))
```

## Opgave 4

```python
#x = 724407.424966 y = 6175719.798486
row = data.loc[data['x']==724407.424966][data['y']==6175719.798486]
name = row.iloc[0,2]
print('Name of monument at location: x = 724407.424966 y = 6175719.798486 is: '+name)
```
