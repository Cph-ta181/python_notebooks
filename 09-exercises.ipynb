```python
import pandas as pd
from itertools import cycle
import matplotlib.pyplot as plt

colors = cycle('bgrcmy')
print(type(colors))

iris_csv = pd.read_csv('../data/iris_data.csv', decimal=',')
iris_csv = iris_csv[['Sepal length', 'Sepal width', 'Species']]
print(iris_csv.head())
print(iris_csv['Species'].unique())

x, y = iris_csv['Sepal length'], iris_csv['Sepal width']
color_list = []
for n in range(len(iris_csv)):
    if iris_csv.iloc[n,2] == 'I. setosa':
        color_list.append('#ff0000')
    elif iris_csv.iloc[n,2] == 'I. versicolor':
        color_list.append('#00ff00')
    else:
        color_list.append('#0000ff')


fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, y, c= color_list, linewidth=0.2)

plt.show()
```
