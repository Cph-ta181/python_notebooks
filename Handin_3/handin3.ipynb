3. Hvad er koefficienten?

5. Hvad vil MPG være når en motor har 200 hestekræfter?

```python
import pandas as pd

data = pd.read_csv('https://think.cs.vt.edu/corgis/datasets/csv/cars/cars.csv')
#print(data)

df = data.loc[data['Identification.Make']=='Honda']
df = df.loc[df['Fuel Information.Fuel Type']=='Gasoline']
df

```

```python
import sklearn
import sklearn.linear_model
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#df.plot.scatter(x=16, y=10)

#Engine Information.Engine Statistics.Horsepower
#Fuel Information.Highway mpg
xs = data['Engine Information.Engine Statistics.Horsepower']
ys = data['Fuel Information.Highway mpg']

xs_reshape = np.array(xs).reshape(-1, 1)
model = sklearn.linear_model.LinearRegression()
model.fit(xs_reshape, ys)
print(f'koefficienten = {model.coef_}')
```

```python
#print(model.intercept_)
plt.scatter(x=df.iloc[:,16],y=df.iloc[:,10])
plt.plot([100, 300],[model.coef_*100+model.intercept_, model.coef_*300+model.intercept_])
```

```python
print(f'Motor med 200 heste krafter vil køre {model.predict([[200]])[0]} MPG')
```
